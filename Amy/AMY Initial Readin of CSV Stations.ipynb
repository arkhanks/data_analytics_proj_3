```python
# Keep only columns needed for our mapping
ev_charging_stations_df = df.loc[:, ['Station Name','Street Address', 'City', 'State', 'ZIP', 'Latitude', 'Longitude', 'EV Level1 EVSE Num',
       'EV Level2 EVSE Num', 'EV DC Fast Count']]
```

```python
ev_charging_stations_df['total EV chargers'] = ev_charging_stations_df[['EV Level1 EVSE Num', 'EV Level2 EVSE Num', 'EV DC Fast Count']].sum(axis=1, skipna=True)
```

```python
ev_charging_stations_df.head()
```

```python
# Replace NaN values with zero in the selected columns
ev_charging_stations_df[['EV Level1 EVSE Num', 'EV Level2 EVSE Num', 'EV DC Fast Count']] = ev_charging_stations_df[['EV Level1 EVSE Num', 'EV Level2 EVSE Num', 'EV DC Fast Count']].fillna(0)

# Display the updated DataFrame
ev_charging_stations_df.head(10)
```

```python
# Save cleaned data to a new CSV file
output_path = r'../Resources/alt_fuel_stations_cleaned.csv'
ev_charging_stations_df.to_csv(output_path, index=False)

output_path
```
